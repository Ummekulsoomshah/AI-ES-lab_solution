```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU activation
def relu(x):
    return np.maximum(0, x)

# Fixed input
X = np.array([[1.0, 2.0]])

# Fixed biases
b1 = np.array([[0.1, 0.2, 0.3]])
b2 = np.array([[0.05, -0.05, 0.1]])

# Randomly vary weights
num_samples = 200
outputs = []

for _ in range(num_samples):
    # Random weights
    W1 = np.random.randn(2, 3)
    W2 = np.random.randn(3, 3)
    
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = relu(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = relu(z2)
    
    outputs.append(a2.flatten())

outputs = np.array(outputs)

# 3D Visualization
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(outputs[:,0], outputs[:,1], outputs[:,2], c='royalblue', s=30)
ax.set_xlabel('Output Neuron 1')
ax.set_ylabel('Output Neuron 2')
ax.set_zlabel('Output Neuron 3')
ax.set_title('Effect of Changing Weights on Output (Fixed Input [1,2])')
plt.show()
```
