```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

X = np.array([[1.0, 2.0]]) 

W1 = np.array([[0.2, -0.4, 0.1],
               [0.7,  0.3, -0.5]])  
b1 = np.array([[0.1, 0.2, 0.3]])    

W2 = np.array([[0.5, -0.1, 0.2],
               [-0.3, 0.8, 0.6],
               [0.4, 0.2, -0.7]])   
b2 = np.array([[0.05, -0.05, 0.1]]) 


z1 = np.dot(X, W1) + b1
a1 = relu(z1)

z2 = np.dot(a1, W2) + b2
a2 = relu(z2)

print("Input:", X)
print("Layer 1 output:", a1)
print("Layer 2 output (final):", a2)
```
