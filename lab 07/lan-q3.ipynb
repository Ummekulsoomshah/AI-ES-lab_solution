```python
import cv2
import numpy as np


cap = cv2.VideoCapture('./car_video.mp4')  


orb = cv2.ORB_create(nfeatures=1000)


while True:
    success, frame = cap.read()
    if not success:
        break


    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)


    keypoints, descriptors = orb.detectAndCompute(gray, None)


    output = cv2.drawKeypoints(frame, keypoints, None, color=(0, 255, 0))


    text = f"Features detected: {len(keypoints)}"
    cv2.putText(output, text, (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)


    cv2.imshow("Feature Detection", output)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()





```
