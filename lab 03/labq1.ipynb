```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Example data (2 features, 6 samples)
data = np.array([
    [1, 200],
    [2, 300],
    [3, 400],
    [4, 500],
    [5, 600],
    [6, 700]
], dtype=float)

# Initialize scalers
std_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# Fit and transform the data
standard_scaled = std_scaler.fit_transform(data)
minmax_scaled = minmax_scaler.fit_transform(data)

# Print results
print("Original Data:\n", data)
print("\nStandardScaler Output:\n", standard_scaled)
print("\nMinMaxScaler Output:\n", minmax_scaled)

# --- Visualization ---
fig, ax = plt.subplots(1, 2, figsize=(10,4))

ax[0].plot(data[:,0], data[:,1], 'bo-', label="Original")
ax[0].set_title("Original Data")

ax[1].plot(standard_scaled[:,0], standard_scaled[:,1], 'go-', label="Standard Scaled")
ax[1].plot(minmax_scaled[:,0], minmax_scaled[:,1], 'ro-', label="MinMax Scaled")
ax[1].legend()
ax[1].set_title("Scaled Data Comparison")

plt.tight_layout()
plt.show()
```
