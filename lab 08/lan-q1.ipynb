```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models
```

```python
# --- Step 1: Upload images for two classes ---
print("ðŸ“¸ Please upload 5â€“10 bright images first (Class 0: bright):")
bright_files = files.upload()

print("ðŸŒ‘ Now upload 5â€“10 dark images (Class 1: dark):")
dark_files = files.upload()

# --- Step 2: Preprocess and extract features ---
def extract_features(image):
    """Resize and compute simple features like brightness and edge density."""
    img = cv2.resize(image, (64, 64))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    brightness = np.mean(gray)
    edges = cv2.Canny(gray, 100, 200)
    edge_density = np.sum(edges > 0) / edges.size
    color_std = np.std(img)
    return np.array([brightness, edge_density, color_std], dtype=np.float32)

def load_images(upload_dict, label):
    features, labels = [], []
    for fn in upload_dict.keys():
        arr = np.frombuffer(upload_dict[fn], np.uint8)
        img = cv2.imdecode(arr, cv2.IMREAD_COLOR)
        if img is not None:
            f = extract_features(img)
            features.append(f)
            labels.append(label)
    return np.array(features), np.array(labels)

# Load and label data
X_bright, y_bright = load_images(bright_files, 0)
X_dark, y_dark = load_images(dark_files, 1)

# Combine data
X = np.vstack([X_bright, X_dark])
y = np.concatenate([y_bright, y_dark])

print(f"âœ… Loaded {len(X)} samples, each with {X.shape[1]} features.")

# --- Step 3: Normalize and split data ---
X = (X - X.mean(axis=0)) / X.std(axis=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Step 4: Build a simple custom neural network ---
model = models.Sequential([
    layers.Input(shape=(3,)),  # 3 input features
    layers.Dense(8, activation='relu'),
    layers.Dense(4, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # output: binary (0 or 1)
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# --- Step 5: Train the model ---
history = model.fit(X_train, y_train, epochs=30, batch_size=4, validation_split=0.2, verbose=1)

# --- Step 6: Evaluate ---
loss, acc = model.evaluate(X_test, y_test)
print(f"ðŸ§  Test Accuracy: {acc*100:.2f}%")

# --- Step 7: Try new image ---
print("\nNow upload one image to test:")
new_file = files.upload()
for fn in new_file.keys():
    arr = np.frombuffer(new_file[fn], np.uint8)
    img = cv2.imdecode(arr, cv2.IMREAD_COLOR)
    features = extract_features(img)
    features = (features - X.mean(axis=0)) / X.std(axis=0)
    pred = model.predict(features.reshape(1, -1))[0][0]
    label = 'Bright (Class 0)' if pred < 0.5 else 'Dark (Class 1)'
    print(f'ðŸ”® Predicted: {label} (probability={pred:.3f})')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f'Predicted: {label}')
    plt.axis('off')
    plt.show()
```
